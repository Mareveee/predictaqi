```python
import numpy as np
from numpy import genfromtxt
import pandas
import matplotlib.pyplot as plot
```

```python
dataset1 = genfromtxt('dataset1.csv',delimiter=',')
output1 = genfromtxt('output1.csv',delimiter=',')
dataset1 = np.array(dataset1,np.float64)
output1 = np.array(output1,np.float64)
print(len(dataset1[23:,0:1]))
print(len(output1))
```

```python
plot.scatter(dataset1[23:,0:1], output1, color = 'red')
```

```python
plot.scatter(dataset1[23:,1:2], output1, color = 'red')
```
