```python
from numpy import genfromtxt
import numpy as np
from sklearn.externals import joblib
from keras.models import load_model
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
```

```python
outputtest = genfromtxt('outputtest.csv',delimiter=',')
model24 = load_model('my.modelLSTM24hr.h5')
test24hr = genfromtxt('dataset.csv',delimiter=',')
test24hr = np.array(test24hr,np.float64)
scaler = joblib.load('my_dope_model24hr.pkl')
test24hr = scaler.transform(test24hr)
test24hr
```

```python
datax=[]
data1 = test24hr[0:len(test24hr)-24,:]
for x in range(len(data1)-11):
    datax.append(data1[0+x:12+x,:])
datax[0]
```

```python
predict=[]
for x in range(len(datax)):
    test24=[]
    test=[]
    test24.append(datax[x])
    test.append(test24)
    predict.append(model24.predict_classes(test))
```

```python
newoutput=[]
for x in range(len(outputtest)-12):
    newoutput.append(outputtest[12+x])
for i in range(len(predict)):
    print(predict[i],newoutput[i])
print(accuracy_score(predict,newoutput))
```
