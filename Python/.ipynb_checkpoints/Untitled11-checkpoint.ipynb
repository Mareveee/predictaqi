```python
from numpy import genfromtxt
import numpy as np
from sklearn.externals import joblib
from keras.models import load_model
from sklearn.metrics import accuracy_score
```

```python
outputtest = genfromtxt('outputtest.csv',delimiter=',')
model1 = load_model('my.modelLSTM1.h5')
test1hr = genfromtxt('dataset.csv',delimiter=',')
test1hr = np.array(test1hr,np.float64)
scaler = joblib.load('my_dope_model.pkl')
test1hr = scaler.transform(test1hr)
test1hr
```

```python
datax=[]
for x in range(0,len(test1hr)-23):
    datax.append(test1hr[15+x:23+x,:])
print(len(datax),len(outputtest))
```

```python
predict=[]
for x in range(len(datax)):
    test1=[]
    test=[]
    test1.append(datax[x])
    test.append(test1)
    predict.append(model1.predict_classes(test))
```

```python
for i in range(len(predict)):
    print(predict[i],outputtest[i])
print(accuracy_score(predict,outputtest))
```
