```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,LSTM
from keras import optimizers
from numpy import genfromtxt
import numpy as np
from keras.utils import to_categorical
```

```python
datasetpm1 = genfromtxt('datasetpm1.csv',delimiter=',')
outputpm1 = genfromtxt('outputpm1.csv',delimiter=',')
datasettestpm = genfromtxt('datasettestpm.csv',delimiter=',')
outputtestpm = genfromtxt('outputtestpm.csv',delimiter=',')
outputtestpm1 = genfromtxt('outputtestpm1.csv',delimiter=',')
datax = []
datay = []
for x in range(len(datasetpm1)-8):
    datax.append(datasetpm1[0+x:8+x,:])
for x in range(len(outputpm1)-8):
    datay.append(outputpm1[8+x])
x_train = np.array(datax,np.float64)
datay = np.array(datay,np.float64)
y_train = to_categorical(datay)
print(x_train)
print(y_train)
```

```python
data_dim = 5
timesteps = 8

model = Sequential()
model.add(LSTM(32,input_shape=(timesteps,data_dim),return_sequences=True))#True = many to many
model.add(LSTM(32, return_sequences=True))  # returns a sequence of vectors of dimension 32
model.add(LSTM(32))
model.add(Dropout(0.3))
model.add(Dense(12,activation='softmax'))
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
model.fit(x_train,y_train,epochs=500);
scores = model.evaluate(x_train, y_train)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
